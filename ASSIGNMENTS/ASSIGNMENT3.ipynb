**6** Find the outliers and replace them outliers

```python
# Find Outliers
# Calculate the z-scores for numeric variables
numeric_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
z_scores = pd.DataFrame((data[numeric_cols] - data[numeric_cols].mean()) / data[numeric_cols].std())
```

```python
# Identify outliers based on z-score threshold (e.g., z-score > 3 or < -3)
outlier_threshold = 3
outliers = data[(z_scores > outlier_threshold) | (z_scores < -outlier_threshold)]
```

```python
# Replace outliers with a specific value (e.g., median)
df_outliers_median = data.copy()
for col in numeric_cols:
    median = df_outliers_median[col].median()
    df_outliers_median.loc[(z_scores[col] > outlier_threshold) | (z_scores[col] < -outlier_threshold), col] = median
```

```python
print("Dataset with outliers replaced with median:\n")
print(df_outliers_median)
```

```python
sns.boxplot(df_outliers_median.area)
```

**7** Check for Categorical columns and perform encoding.

```python
# Check for Categorical Columns
from sklearn.preprocessing import LabelEncoder

categorical_cols = data.select_dtypes(include=['object']).columns
print("Categorical Columns:\n", categorical_cols)
```

```python
# Check for Categorical Columns
categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'furnishingstatus']
categorical_df = data[categorical_cols]
```

```python
# Perform Encoding
# One-Hot Encoding for 'yes/no' columns
one_hot_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning']
one_hot_encoded_df = pd.get_dummies(data[one_hot_cols], drop_first=True)
```

```python
# Label Encoding for remaining categorical column
label_encoder = LabelEncoder()
label_encoded_df = data.copy()
label_encoded_df['furnishingstatus'] = label_encoder.fit_transform(data['furnishingstatus'])
```

```python
# Concatenate the encoded dataframes with the original dataframe
encoded_df = pd.concat([data.drop(categorical_cols, axis=1), one_hot_encoded_df, label_encoded_df['furnishingstatus']], axis=1)
```

```python
data=encoded_df
```
